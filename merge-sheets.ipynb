## Imports

```python
import pandas as pd
```

## Define the path to the source Excel file containing multiple sheets

```python
input_file = r"D:\merged_excel.xlsx"
```

# Define the path for the generated output file

```python
output_file = r"D:\merged_output.xlsx"
```

# Load the Excel file to access its sheet structure

```python
excel_file = pd.ExcelFile(input_file)
```

# Initialize a list to store dataframes from each sheet

```python
all_data = []
```

# Iterate through every sheet name in the workbook

```python
for sheet_name in excel_file.sheet_names:
    # Read the data from the current sheet
    df = pd.read_excel(input_file, sheet_name=sheet_name)
    
    # Insert a new column at the start to track the source sheet name
    df.insert(0, "Sheet_Name", sheet_name)
    
    # Append the dataframe to the master list
    all_data.append(df)

# Concatenate all dataframes from the list into a single master dataframe
merged_df = pd.concat(all_data, ignore_index=True)

# Export the merged dataframe to a new Excel file without including the index
merged_df.to_excel(output_file, index=False)

# Notify the user that the process is complete
print("File successfully created:")
print(output_file)
```
