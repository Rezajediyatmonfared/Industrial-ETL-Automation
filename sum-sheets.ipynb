```python
import os
import pandas as pd

# مسیر پوشه فایل‌های اکسل
folder_path = r"D:\5-ALL"

# مسیر فایل خروجی
output_file = os.path.join(folder_path, "merged_excel.xlsx")

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:

    for file_name in os.listdir(folder_path):

        if file_name.endswith((".xlsx", ".xls")):

            file_path = os.path.join(folder_path, file_name)

            try:
                # خواندن همه شیت‌ها
                excel_file = pd.ExcelFile(file_path)

                for sheet in excel_file.sheet_names:

                    df = pd.read_excel(file_path, sheet_name=sheet)

                    # ساخت نام شیت (نام فایل + نام شیت)
                    sheet_name = f"{os.path.splitext(file_name)[0]}_{sheet}"
                    sheet_name = sheet_name[:31]  # محدودیت اکسل

                    df.to_excel(writer, sheet_name=sheet_name, index=False)

                print(f"✅ پردازش شد: {file_name}")

            except Exception as e:
                print(f"❌ خطا در فایل {file_name}: {e}")

print(f"\n✅ فایل نهایی ساخته شد:")
print(output_file)
```
