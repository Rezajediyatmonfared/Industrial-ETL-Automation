## imports

```python
import os
import pandas as pd
```

# Path to the folder containing Excel files

```python
folder_path = r"D:\5-ALL"
```

# Path to the output file

```python
output_file = os.path.join(folder_path, "merged_excel.xlsx")
```

## Create a new Excel workbook for writing the merged sheets

```python
with pd.ExcelWriter(output_file, engine="openpyxl") as writer:

    # Loop through all files in the target folder
    for file_name in os.listdir(folder_path):

        # Process only Excel files
        if file_name.endswith((".xlsx", ".xls")):

            file_path = os.path.join(folder_path, file_name)

            try:
                # Load the Excel file and read all sheet names
                excel_file = pd.ExcelFile(file_path)

                # Loop through each sheet inside the current Excel file
                for sheet in excel_file.sheet_names:

                    # Read the current sheet into a DataFrame
                    df = pd.read_excel(file_path, sheet_name=sheet)

                    # Build a unique sheet name using file name + sheet name
                    sheet_name = f"{os.path.splitext(file_name)[0]}_{sheet}"

                    # Excel sheet names have a 31-character limit
                    sheet_name = sheet_name[:31]

                    # Write the DataFrame to the output workbook as a new sheet
                    df.to_excel(writer, sheet_name=sheet_name, index=False)

                print(f"✅ Processed: {file_name}")

            except Exception as e:
                print(f"❌ Error in file {file_name}: {e}")

print(f"\n✅ Final file created:")
print(output_file)
```
